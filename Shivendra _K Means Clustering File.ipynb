# K Means Clustering

k- means clustering is one of the simplest algorithm and popular unsupervised machine learning algorithm .you will define a target number k, which refers to number of centroidsyou need in the dataset

A centroid is the imaginary or real location representing the center of the cluster. Every data point is allocated to each of the clusters through reducing the in-cluster sum of squares. In other words, the K-means algorithm identifies k number of centroids, and then allocates every data point to the nearest cluster, while keeping the centroids as small as possible. The means in the K-means refers to averaging of the data; that is, finding the centroid

# Scatter plot - 
A scatter plot in the context of k-means clustering is a type of visualization used to represent the data points and their assigned clusters. It typically shows the data points in a two-dimensional space, with each point plotted based on its features (e.g., x and y coordinates) and colored or marked according to the cluster it belongs to

```python

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs #to generate synthetic data
from sklearn.cluster import KMeans # to perform k-means clustering.

# Generate synthetic data
#n_samples specifies the total number of data points
# centers specifies the number of clusters.
#cluster_std controls the standard deviation of each cluster.

X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Apply K-means clustering
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)#Fit the model to the data using the fit method.
y_kmeans = kmeans.predict(X)#Predict the cluster labels for each data point using the predict method
centroids = kmeans.cluster_centers_ #Obtain the coordinates of the centroids using the cluster_centers_ attribute

# Visualize the clustering
#Scatter plot the data points (X[:, 0] and X[:, 1]) using their first and second features
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
#Color the data points based on their assigned cluster labels (c=y_kmeans).
#Plot the centroids as red circles (plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', s=200, color='red', alpha=0.7, label='Centroids')).
#Add labels and a legend to the plot for clarity.
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', s=200, color='red', alpha=0.7, label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

#xticklabels=['X1', 'X2']: This parameter sets the labels for the x-axis of the heatmap, indicating the dimensions/features of the data (in this case, 'X1' and 'X2').

#yticklabels=['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4']: This parameter sets the labels for the y-axis of the heatmap, representing each cluster (in this case, 'Cluster 1', 'Cluster 2', etc.).

```

# Heat Map
In the context of k-means clustering, a heatmap is a graphical representation that visualizes the centroids of the clusters and their relative positions in the feature space. It provides a clear and intuitive way to understand the distribution of the clusters and the characteristics of the data points within each cluster.

# Cluster Centroids:
After performing k-means clustering on the dataset, the algorithm assigns each data point to one of the k clusters and calculates the centroid of each cluster. A centroid represents the mean position of all the data points in the cluster.

```python
 #Create a heatmap
    
#kmeans.cluster_centers_: This contains the coordinates of the centroids of each cluster obtained from the k-means clustering algorith
plt.figure(figsize=(8, 6))
sns.heatmap(kmeans.cluster_centers_, annot=True, cmap='YlGnBu', fmt='.2f', cbar=True, xticklabels=['X1', 'X2'], yticklabels=['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4'])
#annot=True: This parameter allows annotations (values) to be displayed within each cell of the heatmap, showing the exact coordinates of the centroids
#cmap='YlGnBu': This parameter sets the colormap for the heatmap. Here, 'YlGnBu' represents a colormap that ranges from yellow to green to blue, with increasing values.
plt.title('Heatmap of Cluster Centers')
plt.xlabel('Feature')
plt.ylabel('Cluster')
plt.show()
```

```python

# Generate sample data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Perform k-means clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plot the data points
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5, edgecolors='k', label='Data Points')

# Plot the cluster centroids
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', c='red', s=200, label='Cluster Centroids')

plt.title('Scatter Chart for K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

# Bar chart
A bar chart for k-means clustering is a type of visualization used to represent the distribution of data points among different clusters after performing the k-means clustering algorithm.

```python
#make_blobs is commonly used in machine learning for testing and prototyping clustering algorithms.
#It allows researchers and practitioners to quickly create datasets with known properties to assess the performance and behavior of 
#clustering algorithms under various conditions.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate sample data
X, _ = make_blobs(n_samples=300, centers=5, cluster_std=0.60, random_state=0)

# Perform k-means clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
labels = kmeans.labels_

# Count the number of data points in each cluster
unique, counts = np.unique(labels, return_counts=True)

# Plot the bar chart
plt.figure(figsize=(8, 6))
plt.bar(unique, counts, color='skyblue')
plt.title('Bar Chart for K-means Clustering')
plt.xlabel('Cluster')
plt.ylabel('Number of Data Points')
plt.xticks(unique)
#The line plt.grid(axis='y', linestyle='--', alpha=0.7) configures the gridlines on a matplotlib plot.
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# A pie chart 
for k-means clustering is a type of visualization used to represent the distribution of data points among different clusters after performing the k-means clustering algorithm.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import numpy as np

# Generating sample data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Performing k-means clustering
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
labels = kmeans.labels_

# Counting the number of data points in each cluster
unique, counts = np.unique(labels, return_counts=True)
#The np.unique function is used to find the unique cluster labels (unique) and their corresponding counts (counts).

# Plotting the pie chart
#Matplotlib's plt.pie function is used to create a pie chart.
#counts represents the sizes of the wedges.
#labels=unique specifies the labels for each wedge (which correspond to cluster labels).
#autopct='%1.1f%%' adds percentage labels to each wedge.
#startangle=140 rotates the start of the pie chart to the specified angle.
#plt.axis('equal') ensures that the pie chart is drawn as a circle.
#Titles and labels are added to the plot for clarity (plt.title)

plt.figure(figsize=(8, 6))
plt.pie(counts, labels=unique, autopct='%1.1f%%', startangle=140)
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Cluster Distribution')
plt.show()
```

# Box Plot
In K-means clustering, a box plot isn't typically used directly within the algorithm. However, it might be employed in the context of evaluating the results or understanding the distribution of data within each cluster.

```python


# Generate sample data
X,_  = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Perform k-means clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
labels = kmeans.labels_

#For each cluster:
#X[labels == i, 0] selects the data points belonging to cluster i and extracts the first feature.
#plt.boxplot is used to create a box plot for each cluster's first feature.
#patch_artist=True fills the boxes with colors for better visualization.
#Titles and labels are added to the plot for clarity (plt.title, plt.xlabel, plt.ylabel)
# Create a box plot for each cluster

plt.figure(figsize=(10, 6))
plt.boxplot([X[labels == i, 0] for i in range(kmeans.n_clusters)], patch_artist=True)
plt.title('Box Plot for K-means Clustering')
plt.xlabel('Cluster')
plt.ylabel('Feature 1')

plt.show()
```

```python


# Generate sample data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Perform k-means clustering
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=10, random_state=0)
kmeans.fit(X)

# Plot clustered data points
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.5)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='o', color='red', s=200, label='Centroids')
plt.title('K-means Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

# Histogram
In K-means clustering, a histogram can be a useful tool for visualizing the distribution of data within each cluster. A histogram is a graphical representation of the frequency distribution of a continuous variable. It consists of a series of bars, where the height of each bar represents the frequency of data points falling within a specific range or bin.

```python
# Generating sample data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Performing k-means clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
labels = kmeans.labels_

# Plotting histograms for each cluster
plt.figure(figsize=(10, 6))
for i in range(kmeans.n_clusters):
    plt.hist(X[labels == i, 0], bins=30, alpha=0.5, label=f'Cluster {i + 0}')
plt.title('Histograms for K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
